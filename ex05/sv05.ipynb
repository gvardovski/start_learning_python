```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt

df = pd.read_csv("MSFT_1hour.csv", index_col="Time")
df.index = pd.to_datetime(df.index)
pf_rsi = {}
for i in ['7','14','21']:
    df[f'RSI_{i}'] = ta.rsi(df['Close'], length=int(i))

    entries = df[f'RSI_{i}'] < 30
    exits = df[f'RSI_{i}'] > 70

    pf = vbt.Portfolio.from_signals(
    entries = entries,
    exits = exits,
    open = df["Open"],
    close = df["Close"],
    high = df["High"],
    low = df["Low"],
    size = df['Volume'],
    size_type ="value",
    fees=0,
    slippage=0,
    init_cash=1000,
    freq ="1hour"
    )
    pf_rsi.update({f'RSI_{i}':pf})
for i in ['7','14','21']:
    pf_rsi[f'RSI_{i}'].stats()
```

```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt

df = pd.read_csv("MSFT_1hour.csv", index_col="Time")
df.index = pd.to_datetime(df.index)
df['RSI_14'] = ta.rsi(df['Close'], length = 14)
df['RSI_7'] = ta.rsi(df['Close'], length = 7)
df['RSI_21'] = ta.rsi(df['Close'], length = 21)
entries = df['RSI_21'] < 30
exits = df['RSI_21'] > 70
pf = vbt.Portfolio.from_signals(
    entries = entries,
    exits = exits,
    open = df["Open"],
    close = df["Close"],
    high = df["High"],
    low = df["Low"],
    size = df['Volume'],
    size_type ="value",
    fees=0,
    slippage=0,
    init_cash=1000,
    freq ="1min"
)
pf.stats()
pf.plot()
```
