```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt

df = pd.read_csv("MSFT_1hour.csv", index_col="Time")
df.index = pd.to_datetime(df.index)
pf_rsi = {}
for i in ['7','14','21']:
    df[f'RSI_{i}'] = ta.rsi(df['Close'], length=int(i))

    entries = df[f'RSI_{i}'] < 30
    exits = df[f'RSI_{i}'] > 70

    pf = vbt.Portfolio.from_signals(
    entries = entries,
    exits = exits,
    open = df["Open"],
    close = df["Close"],
    high = df["High"],
    low = df["Low"],
    size = df['Volume'],
    size_type ="amount",
    fees=0,
    slippage=0,
    init_cash=1000,
    freq ="1hour"
    )
    pf_rsi.update({f'RSI_{i}':pf})
```

```python
pf_rsi['RSI_14'].stats()
```

```python
pf_rsi['RSI_14'].plot()
```

```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt

df = pd.read_csv("MSFT_1hour.csv", index_col="Time")
df.index = pd.to_datetime(df.index)
df['Position'] = 0
for i in ['50','200']:
    df[f'SMA_{i}'] = ta.sma(df['Close'], length=int(i))
df = df.loc(df['SMA_200'].notnull())
df.loc[df['SMA_50'] > df['SMA_200'], 'Position'] = 1
df.fillna(0, inplace=True)

entries = df['Position'] > 0
exits = df['Position'] == 0
pf = vbt.Portfolio.from_signals(
    entries = entries,
    exits = exits,
    open = df["Open"],
    close = df["Close"],
    high = df["High"],
    low = df["Low"],
    size = df['Position'],
    size_type ="amount",
    fees=0,
    slippage=0,
    init_cash=1000,
    freq ="1hour"
    )
```

```python
pf.stats()
```

```python
pf.plot()
```
