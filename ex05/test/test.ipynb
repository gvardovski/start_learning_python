```python
import pandas as pd
import vectorbt as vbt

"""
    Long Entry (buy): time == 9:00
    Long exits (exit from buy): time == 12:00
    Position size: 100 units.
"""
df = pd.read_csv("data/EURUSD_1min.csv")
df
df['positions'] = 0.0
df = df.set_index('Time')
df.index = pd.to_datetime(df.index)
df = df[:len(df) // 4]
filtered_df = df.copy().between_time("9:00", "12:00")
df.loc[filtered_df.index, 'positions'] = 100
```

```python
entries = df['positions'] > 0
exits = df['positions'] == 0
exits
```

```python
pf = vbt.Portfolio.from_signals(
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close'],
    entries=entries,
    exits=exits,
    size=df['positions'],
    size_type="amount",
    fees=0,
    slippage=0,
    init_cash=1000,
    freq="1min"
)
```

```python
pf.stats()
```

```python
pf.plot()
```
